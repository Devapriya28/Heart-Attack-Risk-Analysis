```python
continuous=['Age','Cholesterol','Income', 'BMI', 'Triglycerides','Blood Pressure','Heart Rate','Sedentary Hours Per Day','Income', 'BMI', 'Triglycerides']

```

```python
categories=['Diabetes','Family History','Smoking','Obesity','Alcohol Consumption','Previous Heart Problems','Medication Use', 'Stress Level','Physical Activity Days Per Week', 'Sleep Hours Per Day', 'Country',
       'Continent', 'Hemisphere', 'Heart Attack Risk']
```

```python
from scipy import stats

#continuous vs continuous (two sample)
def continuous_vs_continuous(col1,col2):
  t=0
  f=0
  for i in range(31):
    samp1=risk_heart[col1].sample(frac=0.03)
    samp2=risk_heart[col2].sample(frac=0.03)
    ttest,pvalue=stats.ttest_ind(samp1,samp2)
    if pvalue<0.02:
      f=f+1
    else:
      t=t+1

  if t>f:
    return True
  else:
    return False
```

```python
continuous_vs_continuous('Cholesterol','Triglycerides')

```

```python
#continuous(col1) vs category(col2)
def continuous_vs_categories(col1,col2):
  group=risk_heart[col2].unique()
  data={}
  for i in group:
    data[i]=risk_heart[col1][risk_heart[col2]==i]
  fvalue,pvalue=stats.f_oneway(*[i for i in data.values()])
  return False if pvalue<0.2 else True
```

```python
continuous_vs_categories('Triglycerides','Smoking')
```

```python
#category vs category
def categories_vs_categories(col1,col2):
    return True if stats.chi2_contingency(pd.crosstab(risk_heart[col1],risk_heart[col2]))[1] < 0.2 else False
```

```python
categories_vs_categories('Previous Heart Problems','Medication Use')
```

```python
final={}
for i in risk_heart.columns:
  final[i]={}
  for j in risk_heart.columns:
    if i == j:
            final[i][j] = 0  # No self-comparison
            continue
    if (i in continuous) and (j in continuous):
      result=continuous_vs_continuous(i,j)
    elif (i in continuous) and (j in categories):
      result=continuous_vs_categories(i,j)
    elif (i in categories) and (j in continuous):
      result=continuous_vs_categories(j,i)
    elif (i in categories) and (j in categories):
      result=categories_vs_categories(i,j)
    else:
            result = False  # Catch all unmatched types

    final[i][j] = int(result)
```

```python
final
```

```python
final_risk_heart=pd.DataFrame(final)
final_risk_heart
```
